```python
import pandas as pd
import os
```

```python
# os.environ["OMP_NUM_THREADS"] = '1'
import warnings
warnings.filterwarnings('ignore')
```

```python
warm_only_df = pd.read_csv('images/dataset__/mean_warm.csv')
cool_only_df = pd.read_csv('images/dataset__/mean_cool.csv')
warm_cool_df = pd.read_csv('images/dataset__/mean_warm_cool.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
warm_only_df['label'] = le.fit_transform(warm_only_df['label'])
corr = warm_only_df.corr()
corr['label'].sort_values(ascending=False)
```

```python
cool_only_df['label'] = le.fit_transform(cool_only_df['label'])
corr = cool_only_df.corr()
corr['label'].sort_values(ascending=False)
```

```python
warm_cool_df['label'] = le.fit_transform(warm_cool_df['label'])
corr = warm_cool_df.corr()
corr['label'].sort_values(ascending=False)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
# normalize
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
# scale
warm_only_df.iloc[:, :-1] = scaler.fit_transform(warm_only_df.iloc[:, :-1])
cool_only_df.iloc[:, :-1] = scaler.fit_transform(cool_only_df.iloc[:, :-1])
warm_cool_df.iloc[:, :-1] = scaler.fit_transform(warm_cool_df.iloc[:, :-1])
```

```python
# warm only
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
X = warm_only_df.iloc[:, :-1]
y = warm_only_df['label']
```

```python
for i in range(1, 30):
    clf = KNeighborsClassifier(n_neighbors=i)
    scores = cross_val_score(clf, X, y, cv=10).mean()
    print(f'KNN {i}: {scores}')
```

```python
for i in range(1, 30):
    clf = DecisionTreeClassifier(max_depth=i)
    scores = cross_val_score(clf, X, y, cv=10).mean()
    print(f'Decision Tree {i}: {scores}')
```

```python
# rf
n_estimators = range(1, 500, 10)
loss = ['gini', 'entropy']
```

```python
for n in n_estimators:
    for l in loss:
        clf = RandomForestClassifier(n_estimators=n, criterion=l)
        scores = cross_val_score(clf, X, y, cv=10).mean()
        print(f'Random Forest {n} {l}: {scores}')
```

```python
kernel = ['linear', 'poly', 'rbf', 'sigmoid']
C = range(1, 9)
gamma = ['scale', 'auto']
```

```python
for k in kernel:
    for c in C:
        for g in gamma:
            clf = SVC(kernel=k, C=c*10, gamma=g)
            scores = cross_val_score(clf, X, y, cv=10).mean()
            print(f'SVM {k} {c} {g}: {scores}')
            
            clf = SVC(kernel=k, C=c, gamma=g)
            scores = cross_val_score(clf, X, y, cv=10).mean()
            print(f'SVM {k} {c} {g}: {scores}')
            
            clf = SVC(kernel=k, C=c*.1, gamma=g)
            scores = cross_val_score(clf, X, y, cv=10).mean()
            print(f'SVM {k} {c} {g}: {scores}')
```
