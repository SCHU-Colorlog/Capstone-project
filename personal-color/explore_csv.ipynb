```python
import pandas as pd
import numpy as np
```

```python
data_root = 'images/dataset__'
```

```python
import os

spring = pd.read_csv(os.path.join(data_root, 'mean_spring.csv'))
summer = pd.read_csv(os.path.join(data_root, 'mean_summer.csv'))
fall = pd.read_csv(os.path.join(data_root, 'mean_fall.csv'))
winter = pd.read_csv(os.path.join(data_root, 'mean_winter.csv'))
```

```python
spring.head()
```

```python
warm = pd.read_csv(os.path.join(data_root, 'mean_warm.csv'))
cool = pd.read_csv(os.path.join(data_root, 'mean_cool.csv'))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
```

```python
plt.hist(summer['face_l_var'], bins=10, alpha=0.5, label='Summer')
plt.hist(winter['face_l_var'], bins=10, alpha=0.5, label='Winter')
plt.title('Face brightness variance')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
```

```python
plt.figure(figsize=(2, 3))
pastel_palette = sns.color_palette("pastel")
sns.swarmplot(x='label', y='S',data=pd.concat([spring.assign(DataFrame='Spring'), 
                                                        fall.assign(DataFrame='Fall')]), 
              dodge=True, palette=pastel_palette)
plt.text(0.1, spring['S'].median(), f'{spring['S'].median()}', fontsize=8, fontweight='bold', ha='center', va='center', color='black')
plt.text(0.9, fall['S'].median(), f'{fall['S'].median()}', fontsize=8, fontweight='bold', ha='center', va='center', color='black')
plt.tick_params(axis='x', labelsize=8)
plt.tick_params(axis='y', labelsize=8)
plt.title('(a) Warm tone S')
plt.savefig('images/dataset__/warm_tone_s.png')
plt.show()
```

```python
wc = pd.concat([spring, fall, summer, winter])
wc = wc.sample(frac=1).reset_index(drop=True)
wc['label'] = wc['label'].apply(lambda x: 'warm' if (x == 'spring' or x == 'fall') else 'cool')
```

```python
wc.label.unique()
```

```python
wc.to_csv('images/dataset__/mean_warm_cool.csv', index=False)
```

```python
plt.figure(figsize=(2, 3))
pastel_palette = sns.color_palette("pastel")
sns.swarmplot(x='label', y='b',data=pd.concat([wc[wc['label']=='warm'].assign(DataFrame='Warm'), 
                                                       wc[wc['label']=='cool'].assign(DataFrame='Cool')]), 
              dodge=True, palette=pastel_palette)

med1 = warm['b'].median()
med1 = round(med1, 2)
med2 = cool['b'].median()
med2 = round(med2, 2)
plt.text(0.1, med1, f'{med1}', fontsize=8, fontweight='bold', ha='center', va='center', color='black')
plt.text(0.9, med2, f'{med2}', fontsize=8, fontweight='bold', ha='center', va='center', color='black')
plt.tick_params(axis='x', labelsize=8)
plt.tick_params(axis='y', labelsize=8)
plt.title('(e) Warm/Cool b')
plt.savefig('images/dataset__/wc_b.png')
plt.show()
```
