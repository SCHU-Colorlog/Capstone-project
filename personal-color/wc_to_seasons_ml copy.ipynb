```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
warm_cool_df = pd.read_csv('images/dataset__/mean_warm_cool.csv')
```

```python
warm_cool_df.head()
```

```python
warm_cool_df['wc_label'] = warm_cool_df['label'].apply(lambda x: 'warm' if (x == 'spring' or x == 'fall')
                                                       else 'cool')
```

```python
warm_cool_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
warm_cool_df['label'] = le.fit_transform(warm_cool_df['label'])
warm_cool_df['wc_label'] = le.fit_transform(warm_cool_df['wc_label'])
```

```python
warm_cool_df.head()
# cool 0 warm 1
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
warm_cool_df.iloc[:, :-2] = scaler.fit_transform(warm_cool_df.iloc[:, :-2])
```

```python
warm_cool_df.head()
```

```python
X = warm_cool_df.iloc[:, :-2]
y_wc = warm_cool_df['wc_label']
```

```python
corr = warm_cool_df.corr()
corr['wc_label'].sort_values(ascending=False)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import cross_val_score
```

## Warm and Cool

```python
n_estimators = range(1, 500, 10)
loss = ['gini', 'entropy']
```

```python
for n in n_estimators:
    for l in loss:
        clf = RandomForestClassifier(n_estimators=n, criterion=l, random_state=42)
        scores = cross_val_score(clf, X, y_wc, cv=5).mean()
        print(f'Random Forest {n} {l}: {scores}')
```

```python
kernel = ['linear', 'poly', 'rbf', 'sigmoid']
C = range(1, 9)
gamma = ['scale', 'auto']
```

```python
for k in kernel:
    for c in C:
        for g in gamma:
            clf = SVC(kernel=k, C=c*10, gamma=g, random_state=42)
            scores = cross_val_score(clf, X, y_wc, cv=5).mean()
            print(f'SVM {k} {c} {g}: {scores}')
            
            clf = SVC(kernel=k, C=c, gamma=g, random_state=42)
            scores = cross_val_score(clf, X, y_wc, cv=5).mean()
            print(f'SVM {k} {c} {g}: {scores}')
            
            clf = SVC(kernel=k, C=c*.1, gamma=g, random_state=42)
            scores = cross_val_score(clf, X, y_wc, cv=5).mean()
            print(f'SVM {k} {c} {g}: {scores}')
```

```python
for i in range(1, 30):
    clf = DecisionTreeClassifier(max_depth=i)
    scores = cross_val_score(clf, X, y_wc, cv=5).mean()
    print(f'Decision Tree {i}: {scores}')
```

```python
for i in range(1, 30):
    clf = KNeighborsClassifier(n_neighbors=i)
    scores = cross_val_score(clf, X, y_wc, cv=5).mean()
    print(f'KNN {i}: {scores}')
```

```python
# best model
wc_model = SVC(kernel='linear', C=1, random_state=42)

from sklearn.model_selection import cross_val_predict

pred = cross_val_predict(wc_model, X, y_wc, cv=5)
```

```python
pred
```

```python
print('Correct:', (pred == y_wc).sum(), 'out of', len(y_wc))
```
