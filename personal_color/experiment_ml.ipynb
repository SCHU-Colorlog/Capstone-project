```python
from joblib import dump, load
```

```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('images/dataset__/mean_shuffled.csv')
```

```python
df.head()
```

```python
df['coolwarm'] = df['season'].apply(lambda x: 0 if x == 'spring' or x == 'fall' else 1)
```

```python
df
```

```python
warm = pd.concat([df[df['season'] == 'spring'], df[df['season'] == 'fall']])
cool = pd.concat([df[df['season'] == 'summer'], df[df['season'] == 'winter']])
```

```python
warm
```

```python
cool
```

```python
warm['season'] = warm['season'].apply(lambda x: 0 if x == 'spring' else 1)
cool['season'] = cool['season'].apply(lambda x:  0 if x == 'summer' else 1)
```

```python
warm
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
df.iloc[:, :-2] = scaler.fit_transform(df.iloc[:, :-2])
```

```python
df.head()
```

```python
warm.iloc[:, :-2] = scaler.fit_transform(warm.iloc[:, :-2])
cool.iloc[:, :-2] = scaler.fit_transform(cool.iloc[:, :-2])
```

```python
X = df.iloc[:, :-2]
y_wc = df['coolwarm']
```

```python
corr = df.corr()
corr['coolwarm'].sort_values(ascending=False)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import cross_val_score
```

## Warm and Cool

```python
n_estimators = range(1, 500, 10)
loss = ['gini', 'entropy']
```

```python
for n in n_estimators:
    for l in loss:
        clf = RandomForestClassifier(n_estimators=n, criterion=l, random_state=42)
        scores = cross_val_score(clf, X, y_wc, cv=5).mean()
        print(f'Random Forest {n} {l}: {scores}')
```

```python
kernel = ['linear', 'poly', 'rbf', 'sigmoid']
C = range(1, 9)
gamma = ['scale', 'auto']
```

```python
for k in kernel:
    for c in C:
        for g in gamma:
            clf = SVC(kernel=k, C=c*10, gamma=g, random_state=42)
            scores = cross_val_score(clf, X, y_wc, cv=5).mean()
            print(f'SVM {k} {c} {g}: {scores}')
            
            clf = SVC(kernel=k, C=c, gamma=g, random_state=42)
            scores = cross_val_score(clf, X, y_wc, cv=5).mean()
            print(f'SVM {k} {c} {g}: {scores}')
            
            clf = SVC(kernel=k, C=c*.1, gamma=g, random_state=42)
            scores = cross_val_score(clf, X, y_wc, cv=5).mean()
            print(f'SVM {k} {c} {g}: {scores}')
```

```python
for i in range(1, 30):
    clf = DecisionTreeClassifier(max_depth=i)
    scores = cross_val_score(clf, X, y_wc, cv=5).mean()
    print(f'Decision Tree {i}: {scores}')
```

```python
for i in range(1, 30):
    clf = KNeighborsClassifier(n_neighbors=i)
    scores = cross_val_score(clf, X, y_wc, cv=5).mean()
    print(f'KNN {i}: {scores}')
```

```python
# best model
wc_model = SVC(kernel='linear', C=1, random_state=42)

dump(wc_model, 'models/warmcool.joblib') 
```
